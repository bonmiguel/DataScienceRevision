```python
# Date of Attempt: 2026-06-16

"""
Task Description:
Create a function that loads a small dataset of sales records, filters it to find only sales from a specific region, 
groups the filtered data by product, and calculates the total sales amount and average quantity for each product. 
Return a dataframe with columns: product, total_sales, and avg_quantity, sorted by total_sales descending.
"""

import pandas as pd

# Generate a small self-contained dataset
data = {
    'product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone'],
    'region': ['SG', 'MY', 'SG', 'SG', 'SG', 'MY', 'MY', 'SG'],
    'sales': [5000, 4000, 3000, 6000, 5000, 4000, 4000, 3000],
    'quantity': [2, 3, 2, 3, 4, 2, 1, 3]
}

df = pd.DataFrame(data)

def solve_challenge(df: pd.DataFrame, region: str) -> pd.DataFrame:
    filtered = df[df['region'] == region].groupby("product").agg({'sales':'sum', 'quantity':'mean'}).reset_index()
    filtered_rename = filtered.rename(columns={'sales':'total_sales', 'quantity': 'avg_quantity'}).sort_values('total_sales', ascending = False)
    return filtered_rename

# Test your function
result = solve_challenge(df, 'SG')
print(result)
```

```python
# Date of Attempt: 2026-06-16

"""
Task Description:
Create a function that merges two dataframes: customer information and order information. 
After merging, filter to find only customers who have placed orders in the last 30 days (from June 16, 2026).
Group by customer region and calculate total order amount and number of orders for each region.
Return a dataframe with columns: region, total_order_amount, and order_count, sorted by total_order_amount descending.
"""

import pandas as pd
from datetime import datetime, timedelta

# Generate customer dataframe
customers_data = {
    'customer_id': [1, 2, 3, 4, 5, 6],
    'customer_name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'region': ['SG', 'MY', 'SG', 'TH', 'MY', 'SG']
}
customers_df = pd.DataFrame(customers_data)

# Generate order dataframe
orders_data = {
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108],
    'customer_id': [1, 2, 1, 3, 4, 5, 6, 3],
    'order_amount': [5000, 3000, 7000, 4000, 6000, 2500, 8000, 3500],
    'order_date': ['2026-06-10', '2026-05-15', '2026-06-12', '2026-06-08', 
                   '2026-04-20', '2026-06-14', '2026-06-15', '2026-05-25']
}
orders_df = pd.DataFrame(orders_data)

def solve_challenge(customers_df: pd.DataFrame, orders_df: pd.DataFrame) -> pd.DataFrame:
    merged_ = pd.merge(customers_df, orders_df, on='customer_id')
    merged_['order_date'] = pd.to_datetime(merged_['order_date'])
    
    cutoff_date = datetime(2026,6,16) - timedelta(days = 30)

    filtered_ = merged_[merged_['order_date'] >= cutoff_date]

    grouped_ = filtered_.groupby("region").agg({'order_amount': 'sum', 'order_id': 'count'}).reset_index()

    final = grouped_.rename(columns={'order_amount': 'total_order_amount','order_id': 'order_count'}).sort_values('total_order_amount', ascending = False)
    return final

# Test your function
result = solve_challenge(customers_df, orders_df)
print(result)
```

```python
# Date of Attempt: 2026-06-26
"""
Build a single function that cleans, transforms, merges, aggregates, and reshapes sales data.

What the function must do:
- Load the provided sample order and customer data into DataFrames.
- Convert order_date to datetime.
- Rename one column for clarity.
- Handle missing values in customer_segment and unit_price.
- Create a revenue column as quantity * unit_price.
- Filter to keep only rows where quantity is at least 2 and order_date is in 2026.
- Merge orders with customers on customer_id.
- Use a custom function or lambda to create a discount_flag column:
  - 'discounted' if unit_price is below 20
  - 'standard' otherwise
- Group by region and compute:
  - total_revenue
  - average_order_value
  - number_of_orders
- Create a pivot table with region as rows, category as columns, and revenue as values.
- Sort the summary by total_revenue descending.
- Return a dictionary with keys:
  - 'summary'
  - 'pivot'
"""

import pandas as pd

orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "customer_id": [1, 2, 1, 3, 4, 2, 5, 3],
    "order_date": [
        "2026-01-05", "2026-01-18", "2025-12-28", "2026-02-11",
        "2026-03-03", "2026-04-20", "2026-05-15", "2026-06-01"
    ],
    "category": ["Books", "Electronics", "Books", "Home", "Books", "Home", "Electronics", "Books"],
    "quantity": [2, 1, 3, 4, 2, 5, 2, 1],
    "unit_price": [10, 250, None, 30, 15, 40, 180, 12]
})

customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5],
    "customer_name": ["Ava", "Ben", "Cara", "Dan", "Elle"],
    "region": ["West", "East", "West", "North", "East"],
    "customer_segment": ["Retail", None, "Wholesale", "Retail", "Retail"]
})

def solve_challenge(orders, customers):
    
    orders['order_date'] = pd.to_datetime(orders['order_date'])
    orders = orders.rename(columns= {'order_date': 'datetime'})

    merged = pd.merge(orders, customers, on= "customer_id")
    merged = merged.copy()
    merged['customer_segment'] = merged['customer_segment'].fillna("Unknown")
    merged['unit_price'] = merged["unit_price"].fillna(merged['unit_price'].median())
    merged['revenue'] = merged['quantity'] * merged['unit_price']

    filtered = merged[(merged['quantity'] >= 2) & (merged['datetime'].dt.year == 2026)].copy()
    filtered['discount_flag'] = filtered['unit_price'].apply(lambda x: "discounted" if x < 20 else "standard")

    summary = filtered.groupby(['region']).agg(total_revenue = ('revenue','sum'),
                                               average_order_value = ('revenue','mean'),
                                               number_of_orders = ('order_id','count')).reset_index().sort_values('total_revenue', ascending = False)

    pivot = filtered.pivot_table(index='region',
                                 columns='category',
                                 values='revenue',
                                 aggfunc='sum',
                                 fill_value=0)
    return {'summary': summary, 'pivot': pivot}

solve_challenge(orders, customers)
```
