# Series and DataFrames
Create a DataFrame, add a column, append rows, and view the result.

```python
import pandas as pd

data = {
    "Day": [1, 2, 3],
    "Amount": [100, 150, 200],
    "Gender": ["Male", "Female", "Others"],
}

df = pd.DataFrame(data, index=["a", "b", "c"])

# Add a new column
df["Discount"] = [10, 15, 20]

# Add new rows
new_rows = pd.DataFrame(
    [
        {"Day": 4, "Amount": 500, "Gender": "Others", "Discount": 25},
        {"Day": 5, "Amount": 234, "Gender": "Male", "Discount": 80},
    ],
    index=["d", "e"],
)

df = pd.concat([df, new_rows])
df
```
